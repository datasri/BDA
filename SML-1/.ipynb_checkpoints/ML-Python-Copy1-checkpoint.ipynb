## 4. Train/Test Split

**Put all the data related to independent variables into X, and target variable into y**

```python
X = diabetes_orig.iloc[:,0:10]
y = diabetes_orig['Y']
X.head(3)
```

```python
y.head(3)
```

**Perform the train/test split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=12)  
# default split is 0.25, but can be changed through test_size = argument.
```

```python
print (X_train.shape)
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
X_train.head(3)
```

```python
X_test.head(3)
```

## 5. Apply a Machine Learning Model on the data

**Train the data**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(X_train[['BMI',['LTG']], y_train)
```

```python
lr.coef_ # get the model coefficients
```

```python
lr.intercept_ # to see the intercept
```

**sklearn doesn't return p values, so use stats model**

```python
cdata = pd.concat([X_train,y_train],1)
cdata.head()
```

```python
import statsmodels.formula.api as smf
lm1 = smf.ols(formula='Y ~ AGE+SEX+BMI+BP+TCOL+LDL+HDL+LTG+GLU', data=pd.concat([X_train,y_train],1)).fit()
```

```python
lm1.pvalues
```

**Make predictions on train data**

```python
ypred_train = lr.predict(X_train)
ypred_train[0:5]
```

**Calculate the score on train data**

```python
lr.score(X_train,y_train)
```

```python
print ('R2score on train data is: ', r2_score(y_train, ypred_train))
```

**Make predictions on test data**

```python
ypred_test = lr.predict(X_test)
ypred_test[0:5]
```

**Calculate score on test data**

```python
print ('R2score on test data is: ', r2_score(y_test, ypred_test))
```

**Plot test set data predictions vs actual values**

```python
plt.plot(y_test,ypred_test, 'o')
plt.xlabel('Y_true')
plt.ylabel('Y_predictions')
x = np.linspace(0,400,100)
plt.plot(x,x)
plt.show()
```

## 6. Fine tune the model with various options including feature engineering

**Polynomial Features**

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

poly = PolynomialFeatures(1)
X_train_poly= poly.fit_transform(X_train)
X_test_poly= poly.fit_transform(X_test)

lr_poly = LinearRegression()

lr_poly.fit(X_train_poly,y_train)
ypred_train_poly = lr_poly.predict(X_train_poly)
r2_score(y_train,ypred_train_poly)
```

```python
ypred_test_poly = lr_poly.predict(X_test_poly)
r2_score(y_test,ypred_test_poly)
```
