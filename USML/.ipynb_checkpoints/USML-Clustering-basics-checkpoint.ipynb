# Unsupervised Learning 

## Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
%matplotlib inline
```

## 1. Data Generation

```python
from sklearn.datasets import make_blobs
```

```python
bdata = make_blobs(n_samples=300, n_features=2, centers=4, cluster_std=2, random_state=10)
```

```python
X = bdata[0]
y = bdata[1]
```

```python
X.shape
```

```python
y.shape
```

## 2. Data Visualization

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow')
plt.show()
```

## 3. K-Means Clustering 

```python
from sklearn.cluster import KMeans
```

### 3.1 How to choose number of clusters

```python
SSE = []
NC = range(1,10)
for k in NC:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    SSE.append(km.inertia_)
plt.plot(NC,SSE)
plt.xlabel('k')
plt.ylabel('SSE')
```

**We can clearly see that for k=4, SSE dramatically decreases**

**So lets run kmeans for k=4**

```python
kmc = KMeans(n_clusters=4, random_state=0)
```

```python
kmc.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmc.labels_, cmap='rainbow')
```

```python
kmc.cluster_centers_
```

```python
kmc.labels_ # similar to km.predict(X)
```

## 4 Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

plt.figure(figsize=(10, 7)) 
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

### 4.1 Creating clusters

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
hc.fit(X)
hc.labels_
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
ax[0].set_title('K Means')
ax[0].scatter(X[:,0],X[:,1],c=kmc.labels_,cmap='rainbow')

ax[1].set_title("Hierarchical")
ax[1].scatter(X[:,0],X[:,1],c=hc.labels_,cmap='coolwarm')

ax[2].set_title("Original")

ax[2].scatter(X[:,0],X[:,1],c=y, cmap='summer')
```

```python
np.c_[X,y]
```

```python
np.c_[X,hc.labels_]
```

```python
kmc.labels_new = []
for i in kmc.labels_:
    if i == 3:
        kmc.labels_new.append(0)
    elif i == 2:
        kmc.labels_new.append(1)
    elif i == 0:
        kmc.labels_new.append(2)
    elif i == 1:
        kmc.labels_new.append(3)
            
np.array(kmc.labels_new)
```

```python
y
```

```python
hc.labels_new = []
for i in hc.labels_:
    if i == 1:
        hc.labels_new.append(0)
    elif i == 0:
        hc.labels_new.append(1)
    elif i == 3:
        hc.labels_new.append(2)
    elif i == 2:
        hc.labels_new.append(3)
            
np.array(hc.labels_new)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,kmc.labels_new)
```

```python
confusion_matrix(y,hc.labels_new)
```
