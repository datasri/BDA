## Step 1. Load the grocery dataset and prepare the dataset

```python
grocery = pd.read_csv('grocery_dataset.txt', delimiter='\t', header=None)
grocery.head(5)
```

**for mlxtend algorithm data needs to be in binominal format, so above data set needs preparation**

```python
# for apriori algorithm, we will be leveraging dataset needs to be in 
# a particular format.
import csv
grocery_items = set()
with open("grocery_dataset.txt") as f:
    reader = csv.reader(f, delimiter=",")
    for i, line in enumerate(reader):
        grocery_items.update(line)
        output_list = list()
with open("grocery_dataset.txt") as f:
    reader = csv.reader(f, delimiter=",")
    for i, line in enumerate(reader):
        row_val = {item:0 for item in grocery_items}
        row_val.update({item:1 for item in line})
        output_list.append(row_val)
grocery_df = pd.DataFrame(output_list)
grocery_df.head(3)
```

**Lets find the most popular itemS**

```python
grocery.sum(0).sort_values(ascending=False).head(10)
```

```python
popular_items = pd.DataFrame(grocery.sum(0).sort_values(ascending=False)).reset_index()
popular_items.rename(columns = {'index':'item', 0:'freq'}, inplace=True)
popular_items['support'] = popular_items['freq']/grocery.shape[0]
popular_items.head(10)
```

```python
grocery.shape[0] # there are in total 
```

**Lets look at some visualization for 10 most popular item**

```python
fig = plt.figure(figsize=(20,10))
plt.bar(popular_items['item'].head(10), popular_items['support'].head(10))
plt.xticks(rotation=90)
plt.show()
```

## Step 2. Short-list frequently occuring items and item sets by choosing a support level

**We will leverage the apriori algorithm which we imported to do this**

```python
freq_items = apriori(grocery, min_support=0.01, use_colnames=True, max_len = None)
print (freq_items.shape)
```

```python
freq_items.head(5)
```

**lets sort the frequently occuring items**

```python
freq_items.sort_values('support', ascending=False).head(20)
```

## Step 3: From the frequently occuring item sets, generate association rules by choosing a metric

```python
arules = association_rules(freq_items, metric='lift', min_threshold=2)
arules.shape
```

```python
arules.sort_values('confidence', ascending=False).iloc[:,:-2]
```
