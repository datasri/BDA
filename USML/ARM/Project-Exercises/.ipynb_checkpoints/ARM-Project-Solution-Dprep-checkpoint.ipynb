### E.3 Preliminary exploration and visualization

**Find top 5 popular items with freq and support**

```python
# Your code
```

```python
popular_items = pd.DataFrame(store.sum(0).sort_values(ascending=False)).reset_index()
popular_items.rename(columns = {'index':'item', 0:'freq'}, inplace=True)
popular_items['support'] = popular_items['freq']/store.shape[0]
popular_items.head(5)
```

## Step 2. Short-list frequently occuring items and item sets by choosing a support level

### E.4 Obtain frequently occuring items and item sets which occur atleast 2% of the time

```python
freq_items = apriori(store, min_support=0.02, use_colnames=True, max_len = None)
print (freq_items.shape)
```

**Sort the frequently occuring items**

```python
freq_items.sort_values('support', ascending=False).head(10) #not in place
```

**to get just itemsets as a list. apriori library gets them as frozen sets. code below converts frozen set to list**

```python
sfreq = freq_items.sort_values('support', ascending=False)
sfreq.reset_index(drop=True, inplace=True)
ll = list(map(tuple,sfreq['itemsets']))
flat_list = [item for sublist in ll for item in sublist]
flat_list[0:10]
```

## Step 3: From the frequently occuring item sets, generate association rules by choosing a metric

### E.5 Generate association rules which have a lift value of atleast 1.5

```python
#Your code
```

```python
arules = association_rules(freq_items, metric='lift', min_threshold=1.5)
arules.shape
```
