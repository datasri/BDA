## 3. K-Means Clustering 

### 3.1 Fit the Kmeans algorithm to the data

```python
#Kmeans algorithm is buit-in sklearn. we just import it and use it
from sklearn.cluster import KMeans
```

**How many clusters should we use??**

**Lets choose 2**

```python
kmc2 = KMeans(n_clusters=2, random_state=0) # choose any arguments for the KMeans algorithm
```

**Fit the model to the training data with chosen number of clusters**

```python
kmc2.fit(X) # fit the algorithm on the data set with independent variables
```

**Predict cluster labels assigned to each data point**

```python
# One we fit the algoritum on X, with .predict method the algorithm assigns each data point (or row) to a particular cluster
kmc2.predict(X)  # the same result is obtained with .labels_ method
```

```python
kmc2.labels_
```

**Even better to visualize the cluster centers**

```python
plt.scatter(X['X0'], X['X1'], c=kmc2.labels_)
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()
```

### 3.2 How to choose number of clusters

```python
SSE = []
NC = range(1,10)
for k in NC:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    SSE.append(km.inertia_)
plt.plot(NC,SSE)
plt.xlabel('k')
plt.ylabel('SSE')
```

**We can clearly see that for k=4, SSE dramatically decreases**

## 4 Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

**With Hierarchical clustering, we use a dendogram to identify optimal # of clusters**

```python
plt.figure(figsize=(10, 7)) 
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

### 4.1 Creating clusters

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
hc.fit(X)
hc.labels_
```

**Compare plots of kmeans and hierarchical clustering with actual cluster labels**

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
ax[0].set_title('K Means')
ax[0].scatter(X['X0'],X['X1'],c=kmc4.labels_,cmap='rainbow')

ax[1].set_title("Hierarchical")
ax[1].scatter(X['X0'],X['X1'],c=hc.labels_,cmap='coolwarm')

ax[2].set_title("Original")
ax[2].scatter(X['X0'],X['X1'],c=y, cmap='summer')
plt.show()
```
